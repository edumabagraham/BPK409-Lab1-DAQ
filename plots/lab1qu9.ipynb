```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/home/edumaba/Public/MPhil_Thesis/Activity Classification/BPK409-Lab1-DAQ/ExampleData/Lab1Qu9.txt', sep='\t', header=None, names=['x', 'y', 'z', 't'])
```

```python
df['t'] = df['t']/1000
df.head()
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['t'], df['x'], label='X-axis')
plt.plot(df['t'], df['y'], label='Y-axis')
plt.plot(df['t'], df['z'], label='Z-axis')
plt.title('Accelerometer Data')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (g)')
plt.legend()
plt.grid()
```

Visualizing the data as shown in the lab manual.
They viewed a section of the signal and I divided the time by 1000.

```python
# Example: filtering for 6.25s to 8s
start_time = 8.1
end_time = 10.0

# Filter indices where time is within the range
mask = (df['t'] >= start_time) & (df['t'] <= end_time)

# Apply the mask to time and all 3 axis signals
time_zoomed = df['t'][mask]
x_zoomed = df['x'][mask]
y_zoomed = df['y'][mask]
z_zoomed = df['z'][mask]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(time_zoomed, x_zoomed, label='x')
plt.plot(time_zoomed, y_zoomed, label='y')
plt.plot(time_zoomed, z_zoomed, label='z')

plt.xlabel('Time [s]')
plt.ylabel('Acc [g]')
plt.title('Lab1 Qu8 Raw Accelerations')
plt.legend()
plt.grid(True)
plt.show()
```
