```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/home/edumaba/Public/MPhil_Thesis/Activity Classification/BPK409-Lab1-DAQ/ExampleData/Lab1Qu9.txt', sep='\t', header=None, names=['x', 'y', 'z', 't'])
```

```python
df.head()
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['t'], df['x'], label='X-axis', color='r')
plt.plot(df['t'], df['y'], label='Y-axis', color='g')
plt.plot(df['t'], df['z'], label='Z-axis', color='b')
plt.title('Accelerometer Data')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (g)')
plt.legend()
plt.grid()
```
